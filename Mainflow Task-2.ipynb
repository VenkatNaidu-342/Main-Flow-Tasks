```python
data2.drop(['num_critic_for_reviews'], axis=1, inplace=True)
data2.columns
```

```python
Q1 = data2.select_dtypes(include='number').quantile(0.25)
Q3 = data2.select_dtypes(include='number').quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
numeric_cols = data2.select_dtypes(include='number').columns
Q1 = data2[numeric_cols].quantile(0.25)
Q3 = data2[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

data2 = data2[~((data2[numeric_cols] < (Q1 - 1.5 * IQR)) | (data2[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
data2
```
